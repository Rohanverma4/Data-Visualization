**Matplotlib**

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
```

```python
## Example
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
##ploting using matplotlib

##plt scatter

plt.scatter(x,y,c='g')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('2D graph')
plt.savefig('graph.png')
```

```python
y=x*x
```

```python
## plt plot
plt.plot(x,y,'g--')
plt.plot(x,y,'go')
```

```python
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2, markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D diagram')
```

```python
## Creating Subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
plt.subplot(2,2,4)
plt.plot(x,y,'yo')
```

```python
x = np.arange(1,11)
y = 3 * x + 5
plt.title('Matplotlib demo')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(x,y)
```

```python
np.pi
```

```python
x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)
plt.title('sine wave form')

#plotting
plt.plot(x,y)
```

```python
x = np.arange(0, 5 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)

plt.plot(x,y_sin)
plt.title('Sine')

plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title('cosine')
```

***Bar plot***

```python
x = [2,8,12]
y = [11,6,9]

x2 = [3,7,11]
y2 = [6,12,21]
```

```python
plt.bar(x,y)
plt.bar(x2,y2,color = 'g')
plt.title('Bar graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
plt.title('histogram')
plt.show()
```

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]

plt.boxplot(data,vert=True,patch_artist=True)
```

***Pie chart***

```python
#Data 
labels = ['Python','C++','Ruby','Java']
sizes = [215, 130, 245, 210]
colors = ['gold', 'green', 'coral', 'yellow']
explode = (0.2,0,0,0)
```

```python
#Plot
plt.pie(sizes,explode,labels,colors,autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
```
